```python
import pandas as pd


df = pd.read_csv('https://raw.githubusercontent.com/gscdit/Breast-Cancer-Detection/refs/heads/master/data.csv')
df.head()
```

```python
from sklearn.model_selection import train_test_split


df.drop(columns=['id', 'Unnamed: 32'], inplace=True)
x_train, x_test, y_train, y_test = train_test_split(df.iloc[:, 1:], df.iloc[:, 0], test_size=0.2, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler


scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
# from sklearn. import LabelEncoder


from sklearn.preprocessing import LabelEncoder
# Applying LabelEncoder to y_train and y_test will convert categorical labels (like 
# strings 'M' and 'B' in your diagnosis column) into numerical labels (like 1 and 0).


enocoder = LabelEncoder()
y_train = enocoder.fit_transform(y_train)
y_test = enocoder.transform(y_test)
y_train     
```

```python
import torch


X_train_tensor = torch.from_numpy(x_train)
X_test_tensor = torch.from_numpy(x_test)
y_train_tensor = torch.from_numpy(y_train)
y_test_tensor = torch.from_numpy(y_test)
X_train_tensor = X_train_tensor.float()
X_test_tensor = X_test_tensor.float()
X_train_tensor.shape, X_test_tensor.shape, y_train_tensor.shape, y_test_tensor.shape
```

```python
import torch.nn as nn


class NN(nn.Module): 
    def __init__(self, num_features):
        super().__init__()
        self.linear = nn.Linear(num_features, 1)
        self.sigmoid = nn.Sigmoid()

    def forward(self, features):
        return self.sigmoid(self.linear(features))
    
    # def loss_fn(self, y_pred, y):
    #     epsilon = 1e-7
    #     y_pred = torch.clamp(y_pred, epsilon, 1 - epsilon)
    #     loss = -(y * torch.log(y_pred) + (1 - y) * torch.log(1 - y_pred)).mean()
    #     return loss

    
```

```python
learning_rate = 0.1
epochs = 25
```

```python
model = NN(X_train_tensor.shape[1])
loss_fn = nn.BCELoss()
optimizer = torch.optim.SGD(model.parameters(), lr=learning_rate)

# batches me divide karna is better

for epoch in range(epochs):
    y_pred = model(X_train_tensor)
    # loss = model.loss_fn(y_pred, y_train_tensor)
    loss = loss_fn(y_pred, y_train_tensor.view(-1, 1).float())
    # loss.backward()

    # with torch.no_grad():
    #     # for param in model.parameters():
    #     #     param -= learning_rate * param.grad           # 4th file k andar use kar sakte hain
    #     if model.linear.weight.grad is not None:
    #         model.linear.weight -= learning_rate * model.linear.weight.grad
    #     if model.linear.bias.grad is not None:
    #         model.linear.bias -= learning_rate * model.linear.bias.grad # type: ignore

    # model.linear.weight.grad.zero_() # type: ignore
    # model.linear.bias.grad.zero_() # type: ignore

    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    print(f"Epoch: {epoch}, Loss: {loss.item()}")
```

```python
with torch.no_grad():
    y_pred = model(X_test_tensor)
    y_pred = (y_pred > 0.9).float()
    acc = (y_pred == y_test_tensor).float().mean()
    print(f"Test Accuracy: {acc:.2f}")
```
